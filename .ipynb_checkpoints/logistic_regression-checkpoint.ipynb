```python
def Line(x, bias=0, slope=1):
    return slope*x+bias
```

```python
# Negative Slope
x = np.array([1, 100, 200])
y = Line(x, slope=-1)
plt.plot(x, y)
plt.show()
```

# E-Commerce ( Pre-Processing Data )

```python
import pandas as pd
```

```python
def get_data():
    df = pd.read_csv('ecommerce_data.csv') # Df = Data Frames    
    data = df.as_matrix()
    
    # Split The Data
    X = data[:, :-1]
    Y = data[:, -1]
    
    # Normalize numerical columns
    X[:, 1] = (X[:, 1] - X[:, 1].mean()) / X[:, 1].std()
    X[:, 2] = (X[:, 2] - X[:, 2].mean()) / X[:, 2].std()
    
    # Categorical Data ( Time 1 | 2 | 3 | 4 )
    N, D = X.shape
    X2 = np.zeros((N, D+3))
    X2[:, 0:(D-1)] = X[:, 0:(D-1)]
    
    for n in range(N):
        t = int(X[n, D-1])
        X2[n, t+D-1] = 1
    
    return X2, Y
```

```python
def get_binary_data():
    X, Y = get_data()
    X2 = X[Y <= 1]
    Y2 = Y[Y <= 1]
    return X2, Y2
```

## Cross Entropy

```python
N = 100
D = 2
```

```python
X = np.random.randn(N, D)
```

```python
X[:50, :] = X[:50, :] - 2*np.ones((50, D))
```

```python
X[50:, :] = X[50:, :] + 2*np.ones((50, D))
```

```python
T = np.array([0]*50 + [1]*50)
```

```python
ones = np.array([[1]*N]).T
```

```python
Xb = np.concatenate((ones, X), axis=1)
```

###### Random Weights

```python
w = np.random.randn(D+1)
w
```

###### Model output

```python
z = Xb.dot(w)
```

###### Lets do Logistic Regression

```python
Y = sigmoid(z)
```

###### Error!

```python
def cross_entropy(T, Y):
    E = 0
    for i in range(N):
        if(T[i] == 1):
            E -= np.log(Y[i])
        else:
            E -= np.log(1 - Y[i])
    return E
```

```python
print(cross_entropy(T, Y))
```

### Visualize the Line Discriminant

```python
plt.scatter(X[:, 0], X[:, 1], c= T, s=100, alpha=0.5)

x_axis = np.linspace(-6, 6, 100)
y_axis = -x_axis
plt.plot(x_axis, y_axis)
plt.show()
```

### Gradient Descent

```python
learning_rate = 0.1
```

```python
for i in range(230049):
    if(i % 10 == 0):
        print(cross_entropy(T, Y))
    
    w += learning_rate * np.dot((T - Y).T, Xb)
    Y = sigmoid(Xb.dot(w))
    
print("Final w", w)
```
